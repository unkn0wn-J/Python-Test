[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
"""정답지 보고 이해 완료"""

numbers = []

while True:
    try :
        변수 = input("숫자를 입력하세요.")
        if 변수 == "done" :
            break
        numbers.append(int(변수))  #변수에 입력 받은 정수(int)입력값을 리스트(numbers)에 전달

    except ValueError:
        print(f"잘못된 값:{변수}. 숫자를 입력해주세요")
````

[문제 4]

사용자 정의 예외 NegativeNumberError를 생성하고, 사용자로부터 숫자를 입력받아 음수일 경우 NegativeNumberError를 발생시키는 프로그램을 작성하세요. 

- 예외 처리 구문을 사용하여 음수 입력 시 "양수를 입력해주세요."라는 메시지를 출력

````python
"""
class , def 부분에서 막혀서 챗지지피티 및 정답지 활용해서 이해했습니다.
"""


class NegativeNumberError(Exception):  # Exception 상속 필요
    def __init__(self, message):
        self.message = message
    
    def __str__(self):
        return self.message


try :
    numbers = int(input("숫자를 입력해주세요"))
    if numbers < 0:
        raise NegativeNumberError("양수를 입력해주세요.")

except NegativeNumberError as e:
    print(e)

except ValueError :
    print("숫자를 입력해주세요.")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
""" 
이건 잘 모르겠어서 정답지 먼저 확인했습니다 이후 이해 완료.
"""


import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
        return result
    return wrapper

@timer_decorator
def example_function():
    sum = 0
    for i in range(1000000):
        sum += i
    return sum

example_function()
````

````python
"""
챗지피티 이용해서 풀었습니다. 혼자 생각해보다가 막히는 부분이 있어서. 
"""

from math import pi

# 원 넓이
def get_circle():
    radius = float(input("원의 반지름을 입력하세요: "))
    return pi * radius * radius

# 삼각형 넓이
def triangle_area():
    base = float(input("삼각형의 밑변을 입력하세요: "))
    height = float(input("삼각형의 높이를 입력하세요: "))
    return (base * height) / 2

# 직육면체 부피
def cuboid_volume():
    width = float(input("직육면체의 가로 길이를 입력하세요: "))
    height = float(input("직육면체의 세로 길이를 입력하세요: "))
    depth = float(input("직육면체의 높이를 입력하세요: "))
    return width * height * depth
````
