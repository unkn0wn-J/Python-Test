```python
def print_info(**all):              # 키워드 가변 매개변수 all을 받아 딕셔너리로 저장
    for key, value in all.items():  # 딕셔너리의 모든 키-값 쌍을 반복
        print(f"{key}: {value}")    # 키와 값을 "키: 값" 형식으로 출력

print_info(name="은성", age=22, gender="남성") # 함수 호출: 여러 키워드 인자를 전달
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):          # 홀수 생성기 함수 정의: 0부터 n-1까지 숫자 중 홀수만 순서대로 반환
    for i in range(n):        # 0부터 n-1까지 반복
        if i % 2 != 0:           # i가 홀수인지 확인
            yield i              # 홀수면 값을 하나씩 생성 (반환)


for num in odd_numbers(10):     # 생성된 홀수를 하나씩 출력 , odd_numbers(10): 0부터 9까지 숫자 중 홀수를 생성
    print(num)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max(n):
    print("최소값: ", min(n))   #내장함수 min을 이용하여 최소값을 구하는 코드
    print("최대값: ", max(n))   #내장함수 max를 이용하여 최대값을 구하는 코드

num1 = int(input("숫자를 입력하세요"))
num2 = int(input("숫자를 입력하세요"))
num3 = int(input("숫자를 입력하세요"))

n = (num1,num2,num3)        #매개변수 n에 입력받은 num1,2,3 값을 넣기 , 입력받은 세 숫자를 튜플로 묶어서 변수 n에 저장

min_max(n)
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

---

```python
def is_even(numbers):
    return numbers % 2 == 0     # number를 2로 나눈 나머지가 0이면 True, 아니면 False 반환

print(is_even(10))          # is_even 함수에 10을 전달 → 10이 짝수인지 판단, True 출력
print(is_even(5))           # is_even 함수에 5를 전달 → 5가 짝수인지 판단, False 출력
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n == 0:
        return 1 # 기본값 반환, 호출한 함수(factorial)로 전달
    else:
        return n * factorial(n-1) # n보다 1 작은 수의 팩토리얼을 계산하고 그 값에 현재 n을 곱해서 n!을 만든 후 반환

# 아래에서 위로 계산되며, factorial(0)에서 반환된 1부터 곱셈 시작
print(factorial(5))     #print(factorial(4))  <- 5 * 24 = 120
print(factorial(4))     #print(factorial(3))  <- 6 * 4 = 24를 위로 반환
print(factorial(3))     #print(factorial(2))  <- 3 * 2 = 6  , 4를 위로  반환
print(factorial(2))     #print(factorial(1))  <- 값이 되는 1이 위로 또 반환 2 * 1 = 2를 위로 반환
print(factorial(1))     #print(factorial(0))  <- 0이 됐을 때 1로 반환 , 1 * 1 = 1 
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(str):
    for string in str:      # str_list 안의 각 문자열을 하나씩 꺼냄 
        print(len(string))   # 꺼낸 문자열의 길이(글자 수)를 출력

print_lengths(["hello","hi","bye"])
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(s1, s2):
    if len(s1) >= len(s2):  # s1이 s2보다 길거나 같으면
        return s1            # s1 반환  
    else:                   # 그렇지 않으면
        return s2            # s2 반환
        
print(longer_string("hello", "bye"))    
print(longer_string("apple", "banana"))   
```
