__[문제 2: 붕어빵 주문 받기]__

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
"""정답지 보고 이해 했음"""
order = {}
bread_type = input("주문할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵): ")
bread_count = int(input("주문할 붕어빵 개수를 입력하세요: "))
order[bread_type] = bread_count
print(f"주문 내역: {order}")

주문=() #<-딕셔너리에 담긴다
붕어빵종류=입력("주문할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵): ") # < 붕어빵종류를 사용자에게 입력 받기
붕어빵수량=정수(입력("주문할 붕어빵 개수를 입력하세요: ")) # < 붕어빵 개수를 정수로 입력 받기
주문[붕어빵종류] = 붕어빵수량  # < 주문이라는 상자에 붕어빵종류가 키가 되었고, 붕어빵 수량은 값이 되어 저장 되었다.
출력(f"주문 내역: {주문}")
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
"""정답지 보고 이해 완료""" """이 표시는 정답지나 챗지피티를 이용해 풀었다는 걸 의미합니다"""
if stock[bread_type] >= bread_count:
    stock[bread_type] -= bread_count
    print(f"{bread_type}{bread_count}개를 팔았습니다.")
else:
    print(f"죄송합니다.{bread_type}재고가 부족합니다.")

if 재고[붕어빵종류] >= 붕어빵수량:  # 만약 재고에 있는 붕어빵 수가 주문한 수량보다 많거나 같으면
    재고[붕어빵종류] -= 붕어빵수량  # 재고에서 주문한 만큼 빼준다
    print(f"{붕어빵종류} {붕어빵수량}개를 팔았습니다.")  # “빵 몇 개 팔았다”라고 보여준다
else:  # 재고가 주문 수량보다 적으면
    print(f"죄송합니다. {붕어빵종류} 재고가 부족합니다.")  # “재고 부족”이라고 안내
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
"""어려워서 정답지 보고 이해 하겠습니다."""
while True: 
    bread_type = input("주문할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵). 종료를 원하시면 '종료'를 입력하세요: ")
    if bread_type == "종료":
        break
    bread_count = int(input("주문할 붕어빵 개수를 입력하세요: "))
    
    if stock[bread_type] >= bread_count:
        stock[bread_type] -= bread_count
        print(f"{bread_type} {bread_count}개를 판매했습니다.")
    else:
        print(f"죄송합니다. {bread_type} 재고가 부족합니다.")
    
    print("현재 붕어빵 재고:")
    for bread, quantity in stock.items():
        print(f"{bread}: {quantity}개")

while True: #계속 반복하기, 손님이 주문하는 동안 계속 진행
    붕어빵종류 =input("주문할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵). 종료를 원하시면 '종료'를 입력하세요: ") #붕어빵종류를 입력 받는다
    if 붕어빵종류 =="종료": #사용자가 "종료"를 입력시
        break           #반복을 멈추고 프로그램 종료
    붕어빵수량 =int(input("주문할 붕어빵 개수를 입력하세요:")) #정수로 붕어빵 개수를 입력 받는다

    if 재고[붕어빵종류] >= 붕어빵수량:      #만약 재고에 있는 붕어빵 수가 붕어빵 주문 수 보다 크거나 같으면
        재고[붕어빵종류] -= 붕어빵수량      #재고에서 주문한 만큼 빼준다
        print(f"{붕어빵종류}{붕어빵수량}개를 판매했습니다")
    
    print("현재 붕어빵 재고:")
    for 붕어빵종류, 수량 in stock.items():  # 재고 딕셔너리에서 붕어빵 종류와 수량을 하나씩 꺼내기
        print(f"{붕어빵종류}: {수량}개")    # 남은 재고를 화면에 보여주기
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
"""어려워서 정답지 보고 이해 하겠습니다."""
while True:
    bread_type = input("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵). 종료를 원하시면 '종료'를 입력하세요: ")
    if bread_type == "종료":
        break
    additional_stock = int(input(f"{bread_type}의 추가할 개수를 입력하세요: "))
    stock[bread_type] += additional_stock
    print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
    print("현재 붕어빵 재고:")
    for bread, quantity in stock.items():
        print(f"{bread}: {quantity}개")

while True: #계속 반복하기
    붕어빵종류 = input ("추가할 붕어빵 종류를 입력하세요 (팥붕어빵, 슈크림붕어빵, 초코붕어빵). 종료를 원하시면 '종료'를 입력하세요: ") #붕어빵 종류를 입력받기
    if 붕어빵종류 == "종료": # 만약 종료를 입력한다면 
        break           # 반복 중지 및 프로그램 종료
    추가재고 = int(input(f"{bread_type}의 추가할 개수를 입력하세요: ")) #추가재고를 정수로 입력 받기
    재고[붕어빵종류] += 추가재고     # 재고에 추가재고를 입력 받은 만큼 더한다
    print(f"{붕어빵종)의 재고가 {추가재고}개 추가되었습니다.") #붕어빵 종류에 추가된 재고 개수를 화면에 보여주기
    for 붕어빵종류, 수량 in 재고.items(): #재고 딕셔너리에서 붕어빵 종류와 그 수량을 하나씩 꺼내서 반복
        print(f"{붕어빵종}: {수량}개")  # 붕어빵종과 수량을 화면에 보여주기
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
"""어려워서 정답지 보고 이해 하겠습니다."""
prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}
sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}

total_sales = sum(sales[bread] * prices[bread] for bread in sales)
print(f"총 매출: {total_sales}원")
```
